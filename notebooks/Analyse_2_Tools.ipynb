```python
import pandas as pd
import matplotlib.pyplot as plt

# Charger le jeu de données
file_path = '../raw_data/data_cleaned_2021.csv'
data = pd.read_csv(file_path)

# Extraire les colonnes de compétences (colonnes de 'python' à 'google_an')
skills_columns = ['Python', 'spark', 'aws', 'excel', 'sql', 'sas', 'keras', 'pytorch', 'scikit', 'tensor', 'hadoop', 'tableau', 'bi', 'flink', 'mongo', 'google_an']

# Convertir les colonnes de compétences en numériques (entiers)
data[skills_columns] = data[skills_columns].apply(pd.to_numeric, errors='coerce').fillna(0).astype(int)

# Calculer le nombre et les pourcentages de mentions de chaque compétence
skill_counts = data[skills_columns].sum()
total_jobs = len(data)
skill_percentages = (skill_counts / total_jobs) * 100

# Créer un DataFrame pour les pourcentages de compétences et trier par pourcentage en ordre décroissant
skill_percentages_df = skill_percentages.reset_index()
skill_percentages_df.columns = ['Skill', 'Percentage']
skill_percentages_df = skill_percentages_df.sort_values(by='Percentage', ascending=False)

# Créer le graphique en barres
plt.figure(figsize=(12, 8))
plt.bar(skill_percentages_df['Skill'], skill_percentages_df['Percentage'], color='skyblue')
plt.xlabel('Tools')
plt.ylabel('Tools in %')
plt.title('Tools mentioned in job descriptions')
plt.xticks(rotation=90)
plt.tight_layout()

# Afficher le graphique en barres
plt.show()
```
