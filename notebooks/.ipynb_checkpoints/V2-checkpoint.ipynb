```python
import spacy
```

```python
# Chargement du modèle spaCy
nlp = spacy.load("en_core_web_sm")
```

```python
# Texte du job description
job_description = """
We are looking for a Data Scientist who is passionate about using data to drive business 
decisions. The ideal candidate should have strong communication skills, be able to work in 
a fast-paced environment, and have experience with machine learning algorithms such as linear
regression and decision trees.
"""
```

```python
# Liste des softskills
softskills_list = ["communication", "adaptability", "problem-solving", "teamwork", 
                   "creativity", "leadership", "emotional intelligence", "time management"]
```

```python
# Fonction pour l'extraction des softskills à partir du texte
def extract_softskills(text, softskills_list):
    doc = nlp(text)
    extracted_softskills = []
    for token in doc:
        if token.text.lower() in softskills_list:
            extracted_softskills.append(token.text.lower())
    return extracted_softskills
```

```python
# Extraction des softskills à partir du job description
extracted_softskills = extract_softskills(job_description, softskills_list)
```

```python
# Softskills extraits
print("Softskills extraits :", extracted_softskills)
```

```python
# Initialiser une liste pour stocker les softskills extraits de chaque job description
all_extracted_softskills = []

# Extraire les softskills de chaque job description
for job_description in data2['Job Description']:
    extracted_softskills = extract_softskills(job_description, softskills_list)
    all_extracted_softskills.append(extracted_softskills)

# Afficher les softskills extraits
all_extracted_softskills[0]
```
